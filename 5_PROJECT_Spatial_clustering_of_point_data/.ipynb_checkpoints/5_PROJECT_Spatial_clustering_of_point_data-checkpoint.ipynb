### Operacje wykonane w ArcGIS Pro
Wczytaj do ArcGIS Pro otrzymany zestaw danych punktowych `(Zestaw12.xlsx)` oraz plik
`osiedla.shp`. Współrzędne osiedli oraz współrzędne zgłoszonych wykroczeń powinny być
przedstawione w jednakowym układzie współrzędnych `ETRS 1989 Poland CS2000 Zone 7`. Aby
zmienić układ współrzędnych użyj funkcji np. Project. 

Wyeksportuj plik shapefile, który zawiera nowe współrzędne (Feature Class To Shapefile). 


#### 1.  Wczytanie potrzebnych pakietów

```r
library(sf) #pakiet do wczytywania pliku .shp
library(dbscan) #pakiet który zawiera funkcje do klasteryzacji
```

#### 2. Wczytanie danych z plików shape
Wczytaj plik z geometrią osiedli wraz z punktami do RStudio (Jupyter R).

```r
osiedla_plik <- st_read("Projekt/osiedla.shp") #wczytanie pliku z osiedlami
points_file <- st_read("Projekt/Mój_projekt/zestaw12_XYTableToPo_Project.shp") #wczytanie pliku z punktami
```

#### 3. Wczytanie współrzędnych z obiektu przestrzennego sf do macierzy

```r
punkty <- st_coordinates(points_file) #służy do wyodrębnienia współrzędnych z obiektu przestrzennego sf, zwraca macierz zawierającą współrzędne każdego punktu w dwoch kolumnach X oraz Y
```

#### 4. Utworzenie mapy z punktami do klasteryzacji

```r
options(repr.plot.width = 20, repr.plot.height = 16) # ustawienie szerokości okna

#funkcja st_geometry(osiedla_plik) służy do wyodrębnienia geometrii z obiektu, main="" dodaje tytuł wykresu, col="" ustawia kolor wypełnienia granic osiedli, border="" ustawia kolor granic osiedli, asp=1 ustawia jednakowe jednostki na obu osiach, xlab = "" i ylab = "" ustawia etykiety dlka osi
plot(st_geometry(osiedla_plik), main = "Mapa wykroczeń na tle granic osiedli w Krakowie", 
     col = "lightgrey", border = "black", asp = 1, xlab = "X", ylab = "Y")
points(punkty[, 1], punkty[, 2], col = "blue", pch = 19, cex = 0.5) #rysuje punkty na wcześniej stworzonym wykresie z osiedlami, punkty[, 1] i punkty[, 2] oznacza odpowiednia pierwszą i drugą kolumne ze współrzędnymi, col="" ustawia kolor punktów, pch= ustawia typ punktu, cex= ustawia wielkość punktów
axis(1)  # dodaje oś X
axis(2)  # dodaje oś Y
```
