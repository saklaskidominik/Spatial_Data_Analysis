#### 5. Zasada działania algorytmu klasteryzacji gęstościowej - DBSCAN

### DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

DBSCAN to algorytm klasteryzacji oparty na analizie gęstości przestrzennej punktów. Jest stosowany do identyfikacji klastrów w danych o złożonych kształtach, uwzględniając punkty o wysokiej gęstości oraz ignorując punkty szumowe. Jego główne cechy to odporność na szum i możliwość wykrywania klastrów o dowolnym kształcie.

#### Główne pojęcia w DBSCAN:

1. **ε (epsilon):** Promień otoczenia, który określa, jak daleko algorytm szuka sąsiadów wokół punktu. To tzw. promień otoczenia lub dystans wyszukiwania.
2. **minPts:** Minimalna liczba punktów, które muszą znajdować się w zasięgu ε, aby dany punkt mógł być uznany za punkt rdzeniowy klastra.
3. **Punkt rdzeniowy:** Punkt, który ma w swoim otoczeniu (w promieniu ε) co najmniej minPts punktów, w tym siebie samego.
4. **Punkt graniczny:** Punkt, który jest osiągalny z punktu rdzeniowego, ale sam nie spełnia wymagań na punkt rdzeniowy (tj. ma mniej niż minPts sąsiadów w promieniu ε).
5. **Punkt szumowy:** Punkt, który nie jest ani rdzeniowy, ani graniczny i nie można go przypisać do żadnego klastra.

#### Zasada Działania:

1. **Inicjalizacja:**
   - Dla każdego punktu w zbiorze danych algorytm oblicza, ilu ma sąsiadów w promieniu ε.

2. **Identyfikacja punktów rdzeniowych:**
   - Punkt, który ma w swoim otoczeniu (w promieniu ε) co najmniej minPts punktów, jest oznaczany jako punkt rdzeniowy.

3. **Tworzenie klastrów:**
   - Punkty rdzeniowe są łączone w klastry. Punkty rdzeniowe, które są w swoim wzajemnym sąsiedztwie (znajdują się w promieniu ε od siebie), są przypisywane do tego samego klastra.
   - Punkty graniczne, które są w promieniu ε od punktu rdzeniowego, ale nie mają minPts sąsiadów, są przypisywane do klastra, w którym znajduje się punkt rdzeniowy.

4. **Punkty szumowe:**
   - Jeśli punkt nie jest punktem rdzeniowym i nie może zostać przypisany do żadnego klastra (ponieważ nie ma odpowiedniej liczby sąsiadów ani nie znajduje się w promieniu ε żadnego punktu rdzeniowego), zostaje oznaczony jako szum (ang. noise).
   - Punkty szumowe to punkty, które nie są ani rdzeniowe, ani graniczne i są traktowane jako punkty odstające.

5. **Przypisywanie etykiet klastrów:**
   - Punkty rdzeniowe oraz punkty graniczne są oznaczane odpowiednią etykietą klastra, tak aby wszystkie punkty w danym klastrze miały tę samą etykietę.
   - Algorytm przypisuje każdemu klastrowi unikalny identyfikator.

#### Uwagi o parametrach ε i minPts:

- **minPts:**
  - Zbyt małe minPts - algorytm wykryje wiele mniejszych klastrów, co może prowadzić do nadmiernej fragmentacji.
  - Zbyt duże minPts spowoduje, że mniejsze klastry zostaną zignorowane, a niektóre punkty mogą zostać oznaczone jako szum.
  - Minimalna wartość minPts powinna być co najmniej równa liczbie wymiarów danych + 1 (dla danych przestrzennych D oznacza liczbę wymiarów). Twórcy algorytmu sugerują minPts = 2 * D.

- **ε (epsilon):**
  - Wartość ε powinna być na tyle duża, aby w otoczeniu każdego punktu rdzeniowego znajdowała się co najmniej określona liczba punktów (minPts).
  - Dobór odpowiedniej wartości ε jest kluczowy dla uzyskania dobrych wyników klasteryzacji. Często wykorzystywane są narzędzia, takie jak kNNdistplot() w R, które pomagają znaleźć tzw. punkt kolana — miejsce na wykresie, które wskazuje odpowiednią wartość ε.
 be large enough to include at least <code>minPts</code> neighbors for a core point.</li>
    <li>Selecting ε is critical and tools like <code>kNNdistplot()</code> are often used to determine the "knee point" in the plot.</li>
</ul>


#### 6. Zalety algorytmu klasteryzacji gęstościowej - DBSCAN

1. **Brak potrzeby określenia liczby klastrów a priori:**

   W przeciwieństwie do metody k-średnich, DBSCAN nie wymaga wcześniejszego określenia liczby klastrów w danych. Dzięki temu można lepiej dopasować algorytm do rzeczywistego rozkładu danych.

2. **Wykrywanie klastrów o dowolnym kształcie:**

   DBSCAN potrafi wykrywać dowolnie ukształtowane klastry, w tym klastry nieregularne oraz klastry otoczone przez inne klastry. Dzięki parametrowi minPts możliwe jest zredukowanie tzw. efektu single-link, czyli połączenia różnych klastrów cienką linią punktów.

3. **Identyfikacja punktów szumowych i odporność na wartości odstające:**

   Algorytm jest w stanie wydzielać szum, tzn. punkty, które nie pasują do żadnego klastra, są traktowane jako punkty odstające. Dzięki temu DBSCAN jest odporny na wartości odstające, co jest szczególnie ważne w analizie danych z szumem.

4. **Prostota parametrów:**

   DBSCAN wymaga ustawienia tylko dwóch parametrów: promienia otoczenia ε oraz minimalnej liczby punktów w klastrze minPts. To sprawia, że algorytm jest relatywnie prosty w konfiguracji, a kolejność punktów w danych nie wpływa znacząco na wyniki klasteryzacji.

5. **Możliwość ustawienia parametrów przez eksperta:**

   Parametry minPts oraz ε mogą być ustawione przez eksperta w danej dziedzinie, co pozwala na lepsze dopasowanie algorytmu do specyficznych danych, jeśli są dobrze zrozumiane.


#### 7. Wady algorytmu klasteryzacji gęstościowej - DBSCAN

1. **Niepełna deterministyczność:**

   Algorytm DBSCAN nie jest całkowicie deterministyczny: punkty graniczne, które są osiągalne z więcej niż jednego klastra, mogą być przypisane do różnych klastrów w zależności od kolejności przetwarzania danych.
   
   Punkty rdzeniowe oraz punkty szumowe są jednak zawsze deterministycznie przypisywane, ale sposób przypisania punktów granicznych może się różnić.

2. **Problemy z metryką odległości w przypadku wysokiej liczby wymiarów:**

   Jakość DBSCAN zależy od miary odległości — najczęściej używaną metryką jest odległość euklidesowa, która w przypadku danych wielowymiarowych może być mało efektywna z powodu tzw. "klątwy wymiarowości" (ang. curse of dimensionality).
   
   Gdy liczba wymiarów jest duża, różnice między odległościami stają się mniej wyraźne, co utrudnia dobór odpowiedniego promienia ε i osłabia skuteczność klasteryzacji.

3. **Niezdolność do klasteryzacji danych o zróżnicowanej gęstości:**

   Algorytm DBSCAN ma trudności z klasteryzacją zbiorów danych o dużych różnicach gęstości. Nie da się wtedy odpowiednio dobrać wspólnych wartości minPts i ε, które działałyby dla wszystkich klastrów.
   Odpowiednia kombinacja parametrów dla jednego klastra może być niewłaściwa dla innych klastrów, co prowadzi do niepełnej lub błędnej klasteryzacji.

4. **Trudność w doborze odpowiedniej wartości ε:**

   Jeśli dane i skala są niedostatecznie zrozumiane, wybór odpowiedniego progu odległości ε może być trudny.
   
   Zbyt duża lub zbyt mała wartość ε może prowadzić do niewłaściwej klasteryzacji — zbyt wielu klastrów, zbyt małej liczby klastrów, albo nawet wszystkich punktów uznanych za szum.


#### 8. Algorytm klasteryzacji gęstościowej - DBSCAN

```r
# Definiowanie parametrów eps i minPts - obie zmienne takiej samej długości
eps_values <- c(250, 375, 290, 375, 350, 420) #eps to maksymalna odległość, w jakiej punkty mogą być połączone w ten sam klaster
minPts_values <- c(4, 4, 5, 5, 6, 6) #minPts - minimalna liczba punktów, które muszą być w promieniu eps, aby dany punkt mógł być uznany za część klastra

par(mfrow = c(3, 2), mar = c(8, 4, 4, 4)) # Ustawienie układu trzech wykresów obok siebie oraz marginesów

# Sprawdzenie, czy długości obu wektorów są równe - każda wartość eps musi mieć odpowiadającą jej wartość minPts
if (length(eps_values) == length(minPts_values)) 
{
  # Iteracja przez różne wartości parametrów eps i minPts
  for (i in seq_along(eps_values)) #seq_along(eps_values) generuje sekwencję liczbową od 1 do długości wektora eps_values
  {
    eps <- eps_values[i] #przypisujemy aktualne wartości z bieżącej iteracji do zmiennej eps 
    minPts <- minPts_values[i] #przypisujemy aktualne wartości z bieżącej iteracji do zmiennej minPts
    
    # Uruchomienie algorytmu DBSCAN
    dbscan_result <- dbscan(punkty, eps = eps, minPts = minPts) #używamy funkcji dbscan() do klasteryzacji algorytmem DBSCAN i wynik przypisujemy do zmiennej dbscan_result
    points_file$cluster <- dbscan_result$cluster # wynik klasteryzacji jest przypisywany do kolumny cluster obiektu points_file
    
    # Rysowanie granic osiedli
    plot(st_geometry(osiedla_plik), col = "lightgrey", border = "black", 
         main = paste(i, ". DBSCAN z parametrami: eps =", eps, "i minPts =", minPts), asp=1, xlab = "X", ylab = "Y") # funkcja st_geometry(osiedla_plik) służy do wyodrębnienia geometrii z obiektu, main="" dodaje tytuł wykresu, col="" ustawia kolor wypełnienia granic osiedli, border="" ustawia kolor granic osiedli, asp=1 ustawia jednakowe jednostki na obu osiach, xlab = "" i ylab = "" ustawia etykiety dlka osi
    
    # Rysowanie punktów oznaczonych jako szum jednakowym kolorem
    szum_idx <- which(points_file$cluster == 0)  #szum ma wartość 0 w `dbscan`, funkcja which() zwraca indeksy punktów, dla których wartość cluster jest równa 0, czyli zwraca wszystkie punkty oznaczone jako szum, przypisujemy wynik do zmiennej szum_idx
    if (length(szum_idx) > 0) #length(szum_idx) oblicza liczbę punktów oznaczonych jako szum, a if sprawdza czy istnieją jakiś szumy
    {
      plot(st_geometry(points_file[szum_idx, ]), col = rgb(110, 110, 110, maxColorValue = 255), 
           add = TRUE, pch = 19, cex = 0.5) # jesli isteniją punkty szumu to je rysujemy, st_geometry(points_file[szum_idx, ]) wybiera wszystkie wiersze w obiekcie points_file, które odpowiadają indeksom w szum_idx, col= okresla color punktów, add= oznacza ze punkty maja zostać dodane do aktualnego wykresu, pch= oznacza typ punktów, cex= ustawia wielkość punktów
    }
    
    # Rysowanie klastrów z różnymi kolorami
    cluster_idx <- which(points_file$cluster != 0) #funkcja which() zwraca indeksy punktów, dla których wartość cluster jest różna od 0, czyli wszystkie należące do klastrów oprócz szumu
  
    # Liczba klastrów (ignorując szum)
    num_clusters <- length(unique(points_file$cluster[cluster_idx])) # oblicza liczbe klastrów, points_file$cluster[cluster_idx] to pobiera wartości kolumny cluster dla punktów któe należą do klastrów, funkcja unique() zwraca unikalne wartości w tym podzbiorze danych, funkcja length() liczy liczbę unikalnych klastrów
      
    # Generowanie odpowiedniej liczby kolorów
    colors <- rainbow(num_clusters) #rainbow(num_clusters) generuje paletę kolorów składającą się z tylu różnych kolorów, ile wynosi num_clusters

      options(repr.plot.width = 20, repr.plot.height = 20) # ustawienie szerokości okna
      
    # Rysowanie klastrów
    plot(st_geometry(points_file[cluster_idx, ]), 
         col = colors[as.numeric(factor(points_file$cluster[cluster_idx]))], 
         add = TRUE, pch = 19, cex = 0.5) #rysujemy pozostałe punkty w klastrach, st_geometry(points_file[cluster_idx, ]) wybiera wszystkie wiersze w obiekcie points_file, które odpowiadają indeksom w cluster_idx,col=ustawia kolory dla punktów, colors[as.numeric(factor(points_file$cluster[cluster_idx]))] przypisuje każdemu punktowi odpowiedni kolor z palety colors, zgodnie z jego przypisaniem do konkretnego klastra, add= oznacza ze punkty maja zostać dodane do aktualnego wykresu, pch= oznacza typ punktów, cex= ustawia wielkość punktów
    
    # Dodanie osi
    axis(1)  # dodaje oś X
    axis(2)  # dodaje oś Y
  }
}
```

#### 10. Zasada działania algorytmu klasteryzacji gęstościowej - HDBSCAN

### HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise)

HDBSCAN to algorytm klasteryzacji bazujący na gęstości przestrzennej punktów. Jest rozwinięciem algorytmu DBSCAN i jego głównym celem jest wykrycie klastrów o zróżnicowanej gęstości.

#### Zasada Działania:

1. **Identyfikacja punktów rdzeniowych:**
   - a) Dla każdego punktu w zbiorze danych sprawdzana jest liczba sąsiadów w jego otoczeniu. 
   - b) Punkt staje się rdzeniem klastra, jeśli w jego promieniu znajduje się co najmniej minPts sąsiadów.

2. **Tworzenie klastrów:**
   - a) Punkty rdzeniowe, które są w zasięgu otoczenia innych rdzeni, są łączone w jeden klaster.
   - b) Punkty, które znajdują się w promieniu rdzenia, ale nie mają wystarczającej liczby sąsiadów, stają się punktami granicznymi przypisanymi do danego klastra.

3. **Tworzenie hierarchii klastrów:**
   - a) Algorytm tworzy hierarchiczną strukturę klastrów, która przedstawia klastery na różnych poziomach szczegółowości i gęstości.

4. **Wydzielenie punktów szumowych:**
   - a) Punkty, które nie spełniają kryteriów rdzeniowych ani granicznych, są oznaczane jako szum i nie przypisywane do żadnego klastra.

5. **Przypisanie punktów do klastrów:**
   - a) Na podstawie hierarchii i stabilności algorytm przypisuje punkty do odpowiednich klastrów i wyznacza strukturę klastrów w zależności od zmienności gęstości.


#### 11. Zalety algorytmu klasteryzacji gęstościowej - HDBSCAN

1. **Brak konieczności ustalania stałej wartości ε:**
   a) HDBSCAN automatycznie analizuje różne poziomy gęstości, co ułatwia klasteryzację danych o zróżnicowanej gęstości.

2. **Wykrywanie klastrów o zmiennych gęstościach:**
   a) Algorytm jest skuteczny w wykrywaniu klastrów o różnej gęstości, automatycznie przeszukując poziomy gęstości i wyznaczając stabilne klastry.

3. **Odporność na szum:**
   a) HDBSCAN jest odporny na punkty szumowe oraz wartości odstające, co poprawia jakość klasteryzacji w danych o niskiej gęstości punktów.

4. **Wykrywanie klastrów o dowolnym kształcie:**
   a) Algorytm potrafi wykrywać klastry o nieregularnych kształtach dzięki dynamicznej analizie gęstości.

5. **Stabilność klastrów:**
   a) HDBSCAN automatycznie wybiera najbardziej stabilne klastry, co daje pewność co do reprezentatywności wyników.

6. **Tylko jeden kluczowy parametr (minPts):**
   a) Algorytm wymaga tylko jednego kluczowego parametru, co ułatwia jego konfigurację i stosowanie.

7. **Dobre działanie w przypadku dużej skali:**
   a) HDBSCAN dobrze radzi sobie z dużymi zbiorami danych, optymalizując klasteryzację bez potrzeby ręcznego dostosowywania.


#### 12. Wady algorytmu klasteryzacji gęstościowej - HDBSCAN

1. **Wysoka złożoność obliczeniowa:**
   a) Algorytm ma wyższą złożoność obliczeniową niż DBSCAN, co może wpływać na czas przetwarzania dużych zbiorów danych.

2. **Skuteczność zależy od parametru minPts:**
   a) Wartość minPts musi być dobrze dobrana, aby algorytm działał efektywnie, co może być wyzwaniem przy niewłaściwej wartości.

3. **Trudność w interpretacji parametru minPts:**
   a) Wybór parametru minPts może być nieintuicyjny i zależy od zrozumienia struktury danych.

4. **Problemy z klastrowaniem danych o silnym szumie:**
   a) Duża ilość punktów szumowych może wpływać na stabilność klastrów i prowadzić do błędnych wyników.

5. **Problemy z klastrowaniem przy ekstremalnych różnicach skal:**
   a) Chociaż HDBSCAN jest efektywny w identyfikacji klastrów o różnych gęstościach, może napotykać trudności w przypadkach, gdzie klastry występują w bardzo różnych rozmiarach. Ekstremalne różnice w skali klastrów mogą wpływać na spójność i stabilność wykrytych klastrów, co czasem skutkuje niejednoznacznymi lub błędnymi wynikami klasteryzacji.

6. **Ograniczona interpretowalność dla niektórych danych:**
   a) Wyniki klasteryzacji mogą być trudne do interpretacji w przypadku danych wielowymiarowych.

7. **Brak deterministyczności w przypadku granicznych punktów:**
   a) Algorytm nie jest w pełni deterministyczny, co może prowadzić do różnego przypisania punktów granicznych.


#### 13. Algorytm klasteryzacji gęstościowej - HDBSCAN

```r
# Parametry dla HDBSCAN
minPts_values_hdbscan <- c(4, 5, 6, 8, 10, 12, 15, 25)  # Testowanie różnych wartości `minPts`, minPts oznacza minimalną liczbę punktów, która musi znaleźć się w zasięgu (gęstość lokalna), aby punkt mógł być uznany za należący do klastra

par(mfrow = c(4, 2), mar = c(8, 4, 4, 4)) # Ustawienie układu trzech wykresów obok siebie oraz marginesów

# Pętla po różnych wartościach `minPts` dla HDBSCAN
for (i in seq_along(minPts_values_hdbscan)) #seq_along(minPts_values_hdbscan) generuje sekwencję liczbową od 1 do długości wektora minPts_values_hdbscan
  {
  minPts_hdbscan <- minPts_values_hdbscan[i] #przypisujemy aktualne wartości z bieżącej iteracji do zmiennej minPts_hdbscan
  
  # Uruchomienie algorytmu HDBSCAN
  hdbscan_result <- hdbscan(punkty, minPts = minPts_hdbscan) #używamy funkcji hdbscan() do klasteryzacji algorytmem HDBSCAN i wynik przypisujemy do zmiennej hdbscan_result
  points_file$hdbscan_cluster <- hdbscan_result$cluster # wynik klasteryzacji jest przypisywany do kolumny hdbscan_cluster obiektu points_file
  
  # Rysowanie granic osiedli
    plot(st_geometry(osiedla_plik), col = "lightgrey", border = "black", 
         main = paste(i, ". HDBSCAN z parametrami: minPts =", minPts_hdbscan), asp=1, xlab = "X", ylab = "Y") # funkcja st_geometry(osiedla_plik) służy do wyodrębnienia geometrii z obiektu, main="" dodaje tytuł wykresu, col="" ustawia kolor wypełnienia granic osiedli, border="" ustawia kolor granic osiedli, asp=1 ustawia jednakowe jednostki na obu osiach, xlab = "" i ylab = "" ustawia etykiety dlka osi
    
  # Rysowanie punktów oznaczonych jako szum jednakowym kolorem
  szum_idx_hdbscan <- which(points_file$hdbscan_cluster == 0)#szum ma wartość 0 w `hdbscan`, funkcja which() zwraca indeksy punktów, dla których wartość hdbscan_cluster jest równa 0, czyli zwraca wszystkie punkty oznaczone jako szum, przypisujemy wynik do zmiennej szum_idx_hdbscan
  if (length(szum_idx_hdbscan) > 0) #length(szum_idx_hdbscan) oblicza liczbę punktów oznaczonych jako szum, a if sprawdza czy istnieją jakiś szumy
  {
    plot(st_geometry(points_file[szum_idx_hdbscan, ]), 
         col = rgb(120, 120, 120, maxColorValue = 255), add = TRUE, pch = 19, cex = 0.5)# jesli isteniją punkty szumu to je rysujemy, st_geometry(points_file[szum_idx_hdbscan, ]) wybiera wszystkie wiersze w obiekcie points_file, które odpowiadają indeksom w szum_idx_hdbscan, col= okresla color punktów, add= oznacza ze punkty maja zostać dodane do aktualnego wykresu, pch= oznacza typ punktów, cex= ustawia wielkość punktów
  }
  
  # Rysowanie klastrów z różnymi kolorami
  cluster_idx_hdbscan <- which(points_file$hdbscan_cluster != 0) #funkcja which() zwraca indeksy punktów, dla których wartość hdbscan_cluster jest różna od 0, czyli wszystkie należące do klastrów oprócz szumu

  # Liczba klastrów (ignorując szum)
  num_clusters_hdbscan <- length(unique(points_file$hdbscan_cluster[cluster_idx_hdbscan]))# oblicza liczbe klastrów, points_file$hdbscan_cluster[cluster_idx_hdbscan] to pobiera wartości kolumny hdbscan_cluster dla punktów któe należą do klastrów, funkcja unique() zwraca unikalne wartości w tym podzbiorze danych, funkcja length() liczy liczbę unikalnych klastrów

  # Generowanie odpowiedniej liczby kolorów
  colors_hdbscan <- rainbow(num_clusters_hdbscan) #rainbow(num_clusters_hdbscan) generuje paletę kolorów składającą się z tylu różnych kolorów, ile wynosi num_clusters_hdbscan
    
    options(repr.plot.width = 20, repr.plot.height = 26) # ustawienie szerokości okna

  # Rysowanie klastrów
  plot(st_geometry(points_file[cluster_idx_hdbscan, ]), 
       col = colors_hdbscan[as.numeric(factor(points_file$hdbscan_cluster[cluster_idx_hdbscan]))], 
       add = TRUE, pch = 19, cex = 0.5)#rysujemy pozostałe punkty w klastrach, st_geometry(points_file[cluster_idx_hdbscan, ] wybiera wszystkie wiersze w obiekcie points_file, które odpowiadają indeksom w cluster_idx_hdbscan, col=ustawia kolory dla punktów, colors_hdbscan[as.numeric(factor(points_file$hdbscan_cluster[cluster_idx_hdbscan]))] przypisuje każdemu punktowi odpowiedni kolor z palety colors_hdbscan, zgodnie z jego przypisaniem do konkretnego klastra, add= oznacza ze punkty maja zostać dodane do aktualnego wykresu, pch= oznacza typ punktów, cex= ustawia wielkość punktów
  
  # Rysowanie osi
  axis(1)
  axis(2)
}
```
